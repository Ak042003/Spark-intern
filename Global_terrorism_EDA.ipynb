```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt 
import seaborn as sb
```

```python
url='https://bit.ly/2TK5Xn5'
data=pd.read_csv(r"C:\Users\Arpan Kumar\Downloads\globalterrorismdb_0718dist.csv",encoding='latin-1')
```

```python
data.head()
```

```python
data.fillna(0)
```

```python
data.columns.values
```

```python
df=data[['iyear','country_txt','region_txt','city','provstate','success','attacktype1_txt','targtype1_txt','natlty1_txt','gname','weaptype1_txt','nkill','nwound']]
df['iyear']
```

```python
plt.figure(figsize = (15,5))
ax=sb.countplot(x='iyear',data=df)
font = {'family': 'calibri',
        'color':  'Black',
        'weight': 'normal',
        'size': 16,
        }
plt.xticks(rotation=90)
plt.xlabel("Year",fontdict=font)
plt.ylabel("Count",fontdict=font)
plt.title("Number of terrorists activities on yearly basis",fontdict=font)
plt.show()
```

```python
plt.figure(figsize = (15,5))
topcountries=df['country_txt'].value_counts().head(10)
ax=sb.barplot(x=topcountries.index,y=topcountries.values)
font = {'family': 'calibri',
        'color':  'Black',
        'weight': 'normal',
        'size': 16,
        }
plt.title("Terrorists activites in various countries",fontdict=font)
plt.xlabel("Countries",fontdict=font)
plt.ylabel("Count",fontdict=font)
ax.set_xticklabels(labels=topcountries.index,rotation=0)
plt.show()
```

```python
plt.figure(figsize = (15,5))

ax=sb.barplot(x=df['region_txt'].value_counts().values,y=df['region_txt'].value_counts().index)
font = {'family': 'calibri',
        'color':  'Black',
        'weight': 'normal',
        'size': 16,
        }
plt.title("Terrorists activites in various regions",fontdict=font)
plt.ylabel("Countries",fontdict=font)
plt.xlabel("Count",fontdict=font)
plt.show()
```

```python
target=data['targtype1_txt'].value_counts().head(10)

plt.figure(figsize = (25,10))
plt.pie(target.values)
plt.legend(target.index,
          title ="Target type",
          loc ="center right",
          bbox_to_anchor =(1, 0, 0.5, 1))
font = {'family': 'sans',
        'color':  'White',
        'weight': 'normal',
        'size': 26,
        }
plt.title("10 most attacked targets",fontdict=font,loc='right')
plt.show()
```

```python
group=data['gname'].value_counts()[1:11]
sum=group.values.sum()
frac=data['gname'].value_counts()[1:11]/sum*100
plt.figure(figsize = (25,10))
exp=(0.1,0.05,0.01,0,0,0,0,0,0,0)
plt.pie(frac,autopct='%1.1f%%',shadow=True,explode=exp)
plt.legend(group.index,
          title ="Terrorist Group",
          loc ="center right",
          bbox_to_anchor =(1, 0, 0.5, 1))
plt.title("Top 10 Terrorist Groups",fontdict=font,loc='right')
plt.show()
```

```python
plt.figure(figsize = (25,10))
plt.pie(df["attacktype1_txt"].value_counts().values,data=df)
plt.legend(df["attacktype1_txt"].value_counts().index,
          title ="Attack type",
          loc ="center right",
          bbox_to_anchor =(1, 0, 0.5, 1))

plt.title("Type of attack",fontdict=font,loc='right')
plt.show()
```
